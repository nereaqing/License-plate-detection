```python
# Libraries
import os

import random

import shutil

from torch import tensor

from ultralytics import YOLO

import xml.etree.ElementTree as ET

from google.colab import drive
drive.mount('/content/drive')
```

```python
# IF IN THE COMPUTER
# absolute_path = 'C:/Users/nerea/OneDrive/Escritorio/UAB/Tercer/License plate project/License-plate-detection/license plates dataset/'
# xml_dir = 'C:/Users/nerea/OneDrive/Escritorio/UAB/Tercer/License plate project/License-plate-detection/license plates dataset/annotations'

# IF IN GOOGLE COLAB
absolute_path = 'drive/MyDrive/license plates/'
xml_dir = 'drive/MyDrive/license plates/annotations'

folder_path = absolute_path + 'labels'
```

```python
for xml_file in os.listdir(xml_dir):
    if xml_file.endswith('.xml'):
        tree = ET.parse(os.path.join(xml_dir, xml_file))
        root = tree.getroot()

        # Extract the info from the XML
        filename = root.find('filename').text
        img_path = absolute_path + 'images/' + filename
        xmin = int(root.find('.//xmin').text)
        ymin = int(root.find('.//ymin').text)
        xmax = int(root.find('.//xmax').text)
        ymax = int(root.find('.//ymax').text)
        width = int(root.find('.//width').text)
        height = int(root.find('.//height').text)
        label = root.find('.//name').text #We do not need this now

        # Store the data in the YOLO format
        # get the center coordinates
        center_x = ((xmax - xmin) / 2 ) + xmin
        center_y = ((ymax - ymin) / 2) + ymin

        # get the size of the whole image and normalize
        normalized_center_x = round(center_x / width, 2)
        normalized_center_y = round(center_y / height, 2)
        normalized_width = round((xmax - xmin) / width, 2)
        normalized_height = round((ymax - ymin) / height, 2)

        # Write .txt file
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        txt_path = os.path.join(folder_path, filename.split('.')[0] + '.txt')

        with open(txt_path, 'w') as file:
            if label == 'licence':
                content = '0 ' + str(normalized_center_x) + ' ' + str(normalized_center_y) + ' ' + str(normalized_width) + ' ' + str(normalized_height)
                file.write(content)


        print('Successfully created', filename)
```
